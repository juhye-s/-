```python
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
import os 

font_family = 'NanumBarunGothic'
plt.rc('font', family='NanumBarunGothic')
sns.set(font=font_family, rc={"axes.unicode_minus": False})
plt.rcParams['axes.unicode_minus'] = False
```

```python
h1 = df1[['교육시설 ', '도서관', '버스정류소', '복지관', '시각장애인협회', '안마원', '주민센터',
       '지하철', '터미널 ', '편의점 ', '약국', '병원', '관할구별 시각장애인 수', 'label']]

h2 = df2[['교육시설', '도서관', '버스정류소', '복지관', '시각장애인협회', '안마원', '주민센터',
       '지하철', '터미널', '편의점', '약국', '병원', '관할구별 시각장애인 수', 'label',
       'smart_label']]

colormap = plt.cm.PuBu
plt.figure(figsize=(10, 8))
plt.title("Person Correlation of Features", y = 1.05, size = 30)
sns.heatmap(h1.astype(float).corr(), linewidths = 0.1, vmax = 1.0,
           square = True, cmap = colormap, linecolor = "white", annot = True, annot_kws = {"size" : 10})
```

```python
c1 = ['교육시설 ', '도서관', '버스정류소', '복지관', '시각장애인협회', '안마원', '주민센터',
       '지하철', '터미널 ', '편의점 ', '약국', '병원']

for i in c1:
    df1[i] = np.where(df1[i] > 1000, 1000, df1[i])

for i in c1:
    df1[i] = np.where(df1[i] > 0, 1000-df1[i], df1[i])

df1.describe()
```

```python
df3 = df1.pow(2)

print(df1.describe())
print(df3.describe())
```

```python
df3['label'] = df1['label']
df3['관할구별 시각장애인 수'] = df1['관할구별 시각장애인 수']

df3.describe()
```

```python
colormap = plt.cm.PuBu
plt.figure(figsize=(10, 8))
plt.title("Person Correlation of Features", y = 1.05, size = 30)
sns.heatmap(df3.astype(float).corr(), linewidths = 0.1, vmax = 1.0,
           square = True, cmap = colormap, linecolor = "white", annot = True, annot_kws = {"size" : 10})
```

```python
import statsmodels.api as sm

x1 = df3.drop('label', axis = 1)

model3=sm.Logit(df3['label'],x1)
results3=model3.fit(method= 'newton')
print(results3.summary())
```

```python
normalization_x1 = (x1 - x1.min()) / (x1.max() - x1.min()) 
normalization_x1.head()
```

```python
model3=sm.Logit(df3['label'],normalization_x1)
results3=model3.fit(method= 'newton')
print(results3.summary())
```
