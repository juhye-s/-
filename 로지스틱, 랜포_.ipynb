```python
import pandas as pd
import numpy as np
```

```python
df=pd.read_csv('부산광역시데이터셋.csv', encoding='euckr')
df.head()
```

```python
df['label'] = np.where(df['스마트 음향신호기'] < 100, 1, 0)
df=df.drop(['음향신호기','스마트 음향신호기'],axis=1, inplace=False)
df=df.drop(['신호등_X좌표','신호등_Y좌표'],axis=1, inplace=False)
df
```

```python
c=['교육시설', '도서관', '버스 정류소','복지관', '시각장애인 협회', '안마원', '주민센터', '지하철역', '터미널']
#더미
for i in c:
    df[i] = np.where(df[i] > 10000, 10000, df[i])
    
df.describe()
```

```python
#라벨인코딩
from sklearn.preprocessing import LabelEncoder

scaler=LabelEncoder()

df['신호등_자치구군']=scaler.fit_transform(df['신호등_자치구군'])
#더미
df['신호등_자치구군']=df['신호등_자치구군'].astype('category')
df=pd.get_dummies(df)
```

```python
from sklearn.model_selection import train_test_split

x_train, x_test, y_train, y_test = train_test_split(df,df['label'], test_size=0.2, random_state=42, \
                                                    stratify=df['label'])

x_train=x_train.drop(['신호등_index','label'],axis=1)
x_test=x_test.drop(['신호등_index','label'],axis=1)
```

```python
#스케일링
from sklearn.preprocessing import MinMaxScaler
scaler=['교육시설', '도서관', '버스 정류소','복지관', '시각장애인 협회', '안마원', '주민센터', '지하철역', '터미널']
min=MinMaxScaler()
min.fit(x_train[scaler])
x_train[scaler]=min.transform(x_train[scaler])
x_test[scaler]=min.transform(x_test[scaler])
```

```python
#데이터분리

x_train, x_valid, y_train, y_valid = train_test_split(x_train, y_train, test_size=0.2,\
                                                      random_state=42, stratify=y_train)
```

```python
from sklearn.linear_model import LogisticRegression
from sklearn.ensemble import RandomForestClassifier
model=RandomForestClassifier()
model1=LogisticRegression()
```

```python
model.fit(x_train, y_train)
model1.fit(x_train,y_train)
pred=pd.DataFrame(model.predict_proba(x_valid))
pred1=pd.DataFrame(model1.predict_proba(x_valid))
```

```python
#모형평가
from sklearn.metrics import roc_auc_score
#랜포 
roc_auc_score(y_valid, pred.iloc[:,1])
```

```python
#로지스틱회귀 
roc_auc_score(y_valid, pred1.iloc[:,1])
```
